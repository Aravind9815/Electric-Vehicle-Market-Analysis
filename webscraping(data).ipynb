```python

car_name = []
availability = []
range_km = []
efficiency = []
weight = []
acceleration = []
battery = []
fastcharge = []
towing = []
cargo = []
price = []
year = []


total_time = time.time()

for i in range(1,10): 
    start_time = time.time()
    url = f"https://ev-database.org/#group=vehicle-group&rs-pr=10000_100000&rs-er=0_1000&rs-ld=0_1000&rs-ac=2_23&rs-dcfc=0_400&rs-ub=10_200&rs-tw=0_3000&rs-ef=100_350&rs-sa=-1_5&rs-w=1000_3500&rs-c=0_5000&rs-y=2010_2030&s=1&p={i}-10"
    page = requests.get(url)
    soup = BeautifulSoup(page.text, "html.parser")

    

    for x in soup.find_all('div',attrs = {"class":"list-item"}):
        car = x.find('a', attrs={'class':'title'})
        ava = x.find('div', attrs={'class':"availability current"})   
        ran = x.find('span', attrs={'class':'erange_real'})
        eff = x.find('span', attrs={'class':'efficiency'})
        we = x.find('span', attrs={'class':'weight_p'})
        acc = x.find('span', attrs={'class':'acceleration_p'})
        batt = x.find('span', attrs={'class':'battery_p'})
        fast = x.find('span', attrs={'class':'fastcharge_speed_print'})
        tow = x.find('span', attrs={'class':'towweight_p'})
        carg = x.find('span', attrs={'class':'cargo'})
        pric = x.find('span', attrs={'class':'country_uk'})
        yea = x.find('span', attrs={'class':'year_from hidden'})
        
        car_name.append(car.text if car else np.nan)
        availability.append("Yes" if ava else "No")
        range_km.append(ran.text if ran else np.nan)
        efficiency.append(eff.text if eff else np.nan)
        weight.append(we.text if we else np.nan)
        acceleration.append(acc.text if acc else np.nan)
        battery.append(batt.text if batt else np.nan)
        fastcharge.append(fast.text if fast else np.nan)
        towing.append(tow.text if tow else np.nan)
        cargo.append(carg.text if carg else np.nan)
        price.append(pric.text if pric else np.nan)
        year.append(yea.text if yea else np.nan)
     
        

    print(f'Page {i} completed in {time.time()-start_time:.2f} seconds')

print("Total Time Completed in seconds", str(time.time()-total_time))
```

```python
df = pd.DataFrame({
    "Car Name": car_name,
    "Availability": availability,
    "Range (km)": range_km,
    "Efficiency (Wh/km)": efficiency,
    "Weight (kg)": weight,
    "Acceleration (0-100s)": acceleration,
    "Battery (kWh)": battery,
    "Fastcharge (kW)": fastcharge,
    "Towing (kg)": towing,
    "Cargo (L)": cargo,
    "Price (€)": price,
    "Year": year,

})
```

```python
df["Brand"] = df["Car Name"].apply(lambda x:x.split()[0])
```

```python
df["Model"] = df["Car Name"].apply(lambda x:" ".join(x.split()[1:]))
```

```python
df.drop(columns=["Car Name"],axis=1,inplace=True)
```

```python
df["Range (km)"]=df["Range (km)"].apply(lambda x :"".join(re.findall(r"\d+",x))).astype(int)
```

```python
df["Efficiency (Wh/km)"]=df["Efficiency (Wh/km)"].apply(lambda x :"".join(re.findall(r"\d+",x))).astype(int)
```

```python
df["Weight (kg)"]=df["Weight (kg)"].apply(lambda x:"".join(re.findall(r"\d+",x))).astype(int)
```

```python
df["Acceleration (0-100s)"] = df["Acceleration (0-100s)"].apply(lambda x:(x.split()[0]).replace(",",".")).astype(float)
```

```python
df["Battery (kWh)"]=df["Battery (kWh)"].apply(lambda x:(x.split()[0]).replace(",",".")).astype(float)
```

```python
df["Fastcharge (kW)"] = df["Fastcharge (kW)"].apply(lambda x:"".join(re.findall(r"\d+",x))).astype(int)
```

```python
df["Towing (kg)"] = (
    df["Towing (kg)"]
    .astype(str)
    .str.findall(r"\d+")
    .str.join("")
    .replace("", "0")
    .astype(int)
)
```

```python
df["Cargo (L)"]=df["Cargo (L)"].apply(lambda x:(x.split()[0]).replace("unknown","0")).astype(float)
```

```python
a = df["Price (€)"].apply(lambda x:x.replace("N/A","0"))
```

```python
df["Price (€)"] = a.apply(lambda x:"".join(re.findall(r"\d+",x))).astype(int)
```
